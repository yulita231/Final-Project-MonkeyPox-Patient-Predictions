### **Soal 5: Fungsi**  
  
Buatlah sebuah fungsi bernama `hitung_diskon` yang menerima dua argumen: `harga_awal` dan `persen_diskon`. Fungsi ini harus mengembalikan harga setelah diskon.

````python
### YOUR CODE HERE ###
def hitung_diskon(harga_awal, persen_diskon):
    harga_diskon = harga_awal * (1 - persen_diskon / 100)
    return harga_diskon
### END OF YOUR CODE ###

harga_baju = 250000
diskon = 15
harga_akhir = hitung_diskon(harga_baju, diskon)
print(f"Harga awal Rp{harga_baju} dengan diskon {diskon}% menjadi Rp{int(harga_akhir)}")
````

##  **Bagian 2: Exploratory Data Analysis (EDA)**

### **Inspect dataset**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = "https://raw.githubusercontent.com/VenkyAdi/EDA---Projects/refs/heads/main/FedEx%20Logistics%20Performance%20Analysis/SCMS_Delivery_History_Dataset.csv"
df = pd.read_csv(data)

# Tampilkan 5 baris pertama dan info singkat
print("Lima baris pertama:")

print("\nInformasi DataFrame:")
print(df.info())

df.head()
````

### **Penanganan missing values**

````python
data = "https://raw.githubusercontent.com/VenkyAdi/EDA---Projects/refs/heads/main/FedEx%20Logistics%20Performance%20Analysis/SCMS_Delivery_History_Dataset.csv"
df = pd.read_csv(data)
print("Missing values sebelum:")
print(df.isnull().sum())

### YOUR CODE HERE ###
# Isi missing values
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
df['Weight (Kilograms)'] = df['Weight (Kilograms)'].fillna(df['Weight (Kilograms)'].mean())
df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])
df['Dosage'] = df['Dosage'].fillna(df['Dosage'].mode()[0])
df['Line Item Insurance (USD)'] = pd.to_numeric(df['Line Item Insurance (USD)'], errors='coerce')
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(df['Line Item Insurance (USD)'].mean())

### END OF YOUR CODE ###

print("\nMissing values setelah diisi:")
print(df.isnull().sum())
````

### **Analisis Univariat**
Peserta bebas memilih kolom yang ingin dianalisis.

````python
print(df['Shipment Mode'].value_counts())

# Membuat plot untuk visualisasi
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Shipment Mode', order=df['Shipment Mode'].value_counts().index)
plt.title('Distribusi Shipment Mode')
plt.xlabel('Mode Pengiriman')
plt.ylabel('Jumlah')
plt.show()
````

### **Analisis Bivariat & Korelasi**
Peserta bebas memilih pasangan kolom.

````python
# Visualisasi Analisis Bivariat
selected_columns = ['Line Item Quantity', 'Line Item Value', 'Pack Price', 'Unit Price', 'Weight (Kilograms)', 'Line Item Insurance (USD)']
sns.pairplot(df[selected_columns].dropna())
plt.suptitle('Bivariate Analysis of Key Numerical Variables')

# Visualisasi Korelasi
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
````

### **Deteksi outlier**
Peserta bebas memilih kolom numerik yang relevan.

````python
existing_columns = [col for col in selected_columns if col in df.columns]
subset_df = df[existing_columns]

subset_df.plot(kind='box', vert=False, figsize=(12, 8))

plt.title('Distribusi Data untuk Kolom Terpilih', fontsize=16)
plt.xlabel('Nilai (Value)', fontsize=12)
````
